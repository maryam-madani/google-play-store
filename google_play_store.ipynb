*حذف کاراکتر$ از ستون قیمت

```python
googleplay_data['Price'] = googleplay_data['Price'].map(lambda x: x.strip('$'))
```

```python
googleplay_data.head()
```

```python
googleplay_data.query("Price!='0'")
```

*تغییر دادن تایپ ستون قیمت جهت بررسی نمودار توزیع آن

```python
googleplay_data["Price"] = googleplay_data["Price"].astype(object).astype(float)
```

```python
googleplay_data.info()
```

*بررسی توزیع ریتینگ

```python
sns.catplot(x='Price',y='Category',data=googleplay_data)
plt.xlim(0, 60)
```

*حذف کاراکتر های خاصی از ستون تعداد نصب ها جهت تبدیل تایپ این ستون

```python
googleplay_data['Installs'] = googleplay_data['Installs'].map(lambda x: x.strip('+'))
#googleplay_data['Installs'] = googleplay_data['Installs'].map(lambda x: x.strip(','))
googleplay_data['Installs'] = googleplay_data['Installs'].str.replace(r'\D', '')
googleplay_data.sample(5)
```

```python
googleplay_data['Installs'] = googleplay_data['Installs'].astype(object).astype(int)
```
